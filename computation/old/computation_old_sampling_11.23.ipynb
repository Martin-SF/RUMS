````python
# EcoMug: energy distribution
reload(cfg)
df_EM_original = pd.read_hdf('/scratch/mschoenfeld/data_hdf/test3/EcoMug_1e6_gaisser_min730_max3e5_60deg_3D.hdf', key='main')
df_EM_sampled   = pd.read_hdf('/scratch/mschoenfeld/data_hdf/test3/EcoMug_1e6_gaisser_weighted_min730_max3e5_60deg_3D.hdf', key='main')

gamma = -cfg.gaisser_gamma
Emax = float(cfg.max_E)
Emin = float(cfg.min_E)



##########################
# Ludwig ansatz
##########################

if (gamma == 1):
    N = np.log(Emax/Emin)
    # N = np.log(f(np.radians(20), 0, Emax)/f(np.radians(20), 0, Emin))
else:
    N = (-1) * (gamma-1)**(-1) * ( Emax**(-(gamma-1)) - Emin**(-(gamma-1)) )
    # N = (-1) * (gamma-1)**(-1) * ( f(np.radians(20), 0, Emax)**(-(gamma-1)) - f(np.radians(20), 0, Emin)**(-(gamma-1)) )

n = cfg.STATISTICS
E = df_EM_sampled['energy'].values

weights = (N / n) * E**(-2.7 + gamma)
# weights = E**(-2.7 - - gamma) *1e6*5  # manual correction



##########################
# tibits ansatz
# https://user-web.icecube.wisc.edu/~jvansanten/docs/misc/index.html
# latex lesbar:
# https://web.archive.org/web/20231020223243/https://user-web.icecube.wisc.edu/~jvansanten/docs/misc/index.html#weighting-tricks
##########################

from scipy.integrate import tplquad
gaisser = lambda theta, phi, E: 0.14*E**(-2.7)*(
                    1    /(1 + 1.1 * E * np.cos(theta) / (115))
                  + 0.054/(1 + 1.1 * E * np.cos(theta) / (850))
                )

muons_per_cm2_per_second = tplquad(gaisser,
    int(Emin)                               , int(Emax),
    lambda E: 0                        , lambda E: 2*np.pi,
    lambda E, phi: np.radians(float(cfg.min_theta)), lambda E, phi: np.radians(float(cfg.max_theta))
    )
print(f"muons_per_cm2_per_second: {muons_per_cm2_per_second[0]:.3e} +- {muons_per_cm2_per_second[1]:.3e}")
p_gen = E**(-gamma) / (muons_per_cm2_per_second[0]) * np.heaviside(E-Emin, 0) * np.heaviside(Emax-E, 0)

N_gen = cfg.STATISTICS
Atotal = 1*np.pi/3**1e-30

weights = Atotal / ( N_gen * p_gen ) * E**(-2.7)

# weights = 1 / weights  # lol



plib.plot_hist(
    df_EM_sampled['energy'], 
    label=f'gaisser_gamma={gamma} with weights (manual normalization 1e6)' ,
    xlog=True,
    binsize=50,
    show_or_multiplot='multi',
    weights=weights,
    histtype='step'
    )

plib.plot_hist(
    df_EM_sampled['energy'], 
    xlog=True,
    label=f'gaisser_gamma={gamma} no weights' ,
    binsize=50,
    show_or_multiplot='multi',
    histtype='step')

plib.plot_hist(
    df_EM_original['energy'], 
    name=f'{cfg.subfolder}EcoMug - energy distribution in GeV',
    label='gaisser' ,
    ylabel = '# of muons',
    xlabel1 = 'E',
    xlabel2 = 'GeV',
    xlog=True,
    binsize=50,
    show_or_multiplot='show',
    savefig=False,
    histtype='step')
````

To find the solid angle subtended by a conical surface with a half-angle (θ) of 60 degrees at the apex of the cone, you use the formula for the solid angle (Ω) subtended by a cone:

$$ Ω = 2π(1 - \cos(θ)) $$

The angle θ must be in radians, not degrees. Since 1 radian equals approximately 57.2958 degrees, you can convert θ from degrees to radians as follows:

$$ θ_{\text{radians}} = θ_{\text{degrees}} \times \frac{π}{180°} $$

For θ = 60 degrees:

$$ θ_{\text{radians}} = 60° \times \frac{π}{180°} = \frac{π}{3} \text{ rad} $$

Now, using the formula for the solid angle:

$$ Ω = 2π(1 - \cos(\frac{π}{3})) $$

Based on the cosine function:

$$ \cos(\frac{π}{3}) = \frac{1}{2} $$

Plugging this value into the formula:

$$ Ω = 2π(1 - \frac{1}{2}) $$

$$ Ω = 2π(\frac{1}{2}) $$

$$ Ω = π \text{ sr} $$

So, the solid angle subtended by a conical surface with a half-angle of 60 degrees is π steradians.

# Calculate Detector Rate

````python
# calculate a muon flux of a given detector at sea level (using gaisser function)
from scipy.integrate import tplquad
from uncertainties import ufloat
reload(cfg)

emin = float(cfg.min_E)
emax = float(cfg.max_E)
thetamin = float(cfg.min_theta)
thetamax = float(cfg.max_theta)

# emin = 1e
# emax = 1e2
# thetamin = 0
# thetamax = 90

f = lambda theta, phi, E: 0.14*E**(-2.7)*(
                    1    /(1 + 1.1 * E * np.cos(theta) / (115))
                  + 0.054/(1 + 1.1 * E * np.cos(theta) / (850))
                ) 

muons_per_x = tplquad(f,
    int(emin)                               , int(emax),
    lambda E: 0                        , lambda E: 2*np.pi,
    lambda E, phi: np.radians(thetamin), lambda E, phi: np.radians(thetamax)
    )

print(f'{muons_per_x[0]*1e4:.3} Muons per second per m^2 (should be around 100 for full integration limits...?)')

detector_area = cfg.artificial_detector_area
muon_rate = muons_per_x[0]*60*60*24*detector_area
muon_rate_u = ufloat(muon_rate, muons_per_x[1])

# print(f'{muons_per_x[0]/scale_factor} +- {muons_per_x[1]/scale_factor} / 1/cm^2*min')
print(f'radius = {cfg.radius_target_circle:.2} cm | artificial_detector_area = {detector_area:.2} cm^2')
print(f'{muon_rate_u:.2} Muons per Day for artificial detector surface (with numerical error)')


# (633.0) of 1e+06 (0.0633)% detector hits | min(E_i) at detector = 709.9 GeV
percent_of_muons_arrived = len(df_pp['point1x'])/cfg.STATISTICS
print(f'{muon_rate*percent_of_muons_arrived:.1f} Muons per Day (The real detectorrate!)')
# muon_rate * percent_of_muons_arrived/cfg.calibrate_which_get_detected
````
